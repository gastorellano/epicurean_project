```python
# Mergear df1 y df2 con outer join
df_merged = pd.merge(df_percapitaingresoCONDAD, df_alquilerCONDADO, on='County', how='outer')

# Mergear el resultado con df3 también con outer join
df_datosEconomicos = pd.merge(df_merged, df_precioviviendaCONDADO, on='County', how='outer')
```

```python
df_datosEconomicos.columns
```

```python
# Eliminar filas duplicadas basadas en la columna 'county'
df_datosEconomicos = df_datosEconomicos.drop_duplicates(subset='County')
```

```python
nuevo_condado = pd.DataFrame({
    'County': ['Acton'],  # Reemplaza 'nuevo_condado' por el nombre real
    'Per capita income': [None],  # Puedes usar un valor realista o None
    'Rent': [None],  # Puedes usar un valor realista o None
    'Property Cost': [None]  # Puedes usar un valor realista o None
})

# Concatenar el DataFrame con el existente
df_datosEconomicos = pd.concat([df_datosEconomicos, nuevo_condado], ignore_index=True)
```

```python
# Función para generar un valor aleatorio cerca de la mediana
def valor_aleatorio_cercano(mediana):
    rango = 0.1 * mediana  # Ajusta el rango según sea necesario (10% de la mediana)
    return np.random.uniform(mediana - rango, mediana + rango)

# Rellenar los valores vacíos con valores aleatorios cercanos a la mediana
for columna in ['Per capita income', 'Rent', 'Property Cost']:
    mediana = df_datosEconomicos[columna].median()  # Calcular la mediana una vez por columna
    for index, row in df_datosEconomicos.iterrows():
        if pd.isna(row[columna]):
            df_datosEconomicos.at[index, columna] = round(valor_aleatorio_cercano(mediana), 2)

```

```python
for columna in ['Per capita income', 'Rent', 'Property Cost']:
    df_datosEconomicos[columna] = df_datosEconomicos[columna].astype(int)
```
