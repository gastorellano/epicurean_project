```python
# Función para normalizar los nombres: eliminar "county" o "region", convertir a minúsculas, quitar espacios extra y eliminar acentos
def normalizar_nombre(nombre):
    nombre = re.sub(r'\b(county|region)\b', '', nombre, flags=re.IGNORECASE)  # Eliminar "county" o "region"
    nombre = nombre.lower()  # Convertir a minúsculas
    nombre = nombre.strip()  # Eliminar espacios en blanco extra
    return nombre

# Aplicar la normalización a la columna 'county_name' en tu DataFrame
Household['county_name_normalizado'] = Household['county_name'].apply(normalizar_nombre)

# También normalizamos la lista de condados
lista_condados_normalizados = [normalizar_nombre(condado) for condado in lista_condados_lower]

# Filtrar el DataFrame usando los condados normalizados
filtered_df = Household[Household['county_name_normalizado'].isin(lista_condados_normalizados)]

# eliminamos la columna normalizada
filtered_df = filtered_df.drop(columns=['county_name_normalizado'])
```

```python
# Aplicar la eliminación de "county" o "region" directamente en la columna 'county_name' del DataFrame filtrado
filtered_df['county_name'] = filtered_df['county_name'].apply(lambda nombre: re.sub(r'\b(county|region)\b', '', nombre, flags=re.IGNORECASE).strip())
```

```python
filtered_df.rename(columns={'county_name': 'County','value': 'Median Household Income'}, inplace=True)
```

```python
df_household = filtered_df.groupby('County', as_index=False).agg({
    'Median Household Income': 'sum'
})
```

***Extraemos y modficimaos datos sobre el la poblacion***

```python
# Ruta del archivo .xlsx
file_path = r'C:\Users\Usuario\Desktop\HENRY\PROYECTO FINAL\CENSO\V2\PoblacionPorCiudad.xlsx'

# Cargar el archivo .xlsx y seleccionar columnas específicas
columnas_necesarias = ['city', 'population']
poblacion = pd.read_excel(file_path, usecols=columnas_necesarias)

# Dividir la columna 'city' en 'city_name' y 'state'
poblacion[['city_name', 'state']] = poblacion['city'].str.split(', ', expand=True)

# Eliminar la palabra "city" de los nombres de las ciudades
poblacion['city_name'] = poblacion['city_name'].str.replace(' city', '', regex=False)

# Crear una lista para almacenar los resultados
resultados = []

# Sumar la población por condado
for condado, ciudades in condados_a_ciudades.items():
    # Filtrar las ciudades que están en el DataFrame original
    ciudades_filtradas = poblacion[poblacion['city_name'].isin(ciudades)]
    poblacion_total = ciudades_filtradas['population'].sum()

    # Agregar el resultado a la lista
    resultados.append({'condado': condado, 'poblacion_total': poblacion_total})

# Crear un nuevo DataFrame con los resultados
df_condados_poblacion = pd.DataFrame(resultados)
```

```python
# Calcular el promedio de la columna 'poblacion_total' excluyendo los ceros
promedio = df_condados_poblacion['poblacion_total'][df_condados_poblacion['poblacion_total'] != 0].mean()

# Definir un rango para el valor aleatorio (por ejemplo, ±20% del promedio)
rango = promedio * 0.2

# Reemplazar los ceros con un valor aleatorio alrededor del promedio
for index in df_condados_poblacion.index:
    if df_condados_poblacion.at[index, 'poblacion_total'] == 0:
        df_condados_poblacion.at[index, 'poblacion_total'] = np.random.uniform(promedio - rango, promedio + rango)
```

```python
# Convertir la columna 'poblacion_total' a tipo entero
df_condados_poblacion['poblacion_total'] = df_condados_poblacion['poblacion_total'].astype(int)
```

```python
df_condados_poblacion.rename(columns={'condado': 'County'}, inplace=True)
```

***Ajustamos valores de Poblacion***

```python
# Paso 1: Calcular el exceso de población
poblacion_actual = df_condados_poblacion['poblacion_total'].sum()
poblacion_objetivo = 39_500_000
diferencia = poblacion_actual - poblacion_objetivo

# Paso 2: Calcular el factor de ajuste
factor_ajuste = diferencia / poblacion_actual

# Paso 3: Aplicar la corrección proporcionalmente a cada condado
df_condados_poblacion['poblacion_total'] = df_condados_poblacion['poblacion_total'] * (1 - factor_ajuste)

# Mostrar el total ajustado y las nuevas poblaciones
poblacion_total_ajustada = df_condados_poblacion['poblacion_total'].sum()
```

```python
# Función para generar un valor aleatorio cerca de la mediana
def valor_aleatorio_cercano(mediana):
    rango = 0.1 * mediana  # Ajusta el rango según sea necesario (10% de la mediana)
    return np.random.uniform(mediana - rango, mediana + rango)

# Rellenar los valores vacíos con valores aleatorios cercanos a la mediana
for columna in ['Per capita income', 'Rent', 'Property Cost','Population','Median Household Income']:
    mediana = df_datosEconomicos[columna].median()  # Calcular la mediana una vez por columna
    for index, row in df_datosEconomicos.iterrows():
        if pd.isna(row[columna]):
            df_datosEconomicos.at[index, columna] = round(valor_aleatorio_cercano(mediana - (mediana * 0.20)), 2)
```
